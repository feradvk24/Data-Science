```python
import pandas as pd
import json
df = pd.read_csv("hf://datasets/maharshipandya/spotify-tracks-dataset/dataset.csv")
```

```python
# Load the JSON file
with open('all.json', 'r') as file:
    json_data = json.load(file)

# Normalize the nested 'data' field
df2 = pd.json_normalize(
    json_data, 
    record_path='data', 
    meta=['date'], 
    errors='ignore'
)
```

```python
df
```

```python
df2
```

```python
df2.date = pd.to_datetime(df2.date)
sorted_df = df2.sort_values(by = ['date','weeks_on_chart', 'this_week'], ascending = [False, False, True])
sorted_df = sorted_df.drop_duplicates(subset = ["song", "artist"], keep = 'first')
sorted_df = sorted_df.sort_values(by = ['date', 'this_week'], ascending = [False, True])
```

```python
df_merged = sorted_df.merge(df, left_on = ['song','artist'], right_on = ['track_name', 'artists'])
```
